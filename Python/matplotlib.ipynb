# matplotlib
## グラフの描画
グラフの表示にはmatplotlib.pyplotモジュールのplot()およびshow()を使用。

```
from matplotlib import pyplot

pyplot.plot(x軸, y軸)
pyplot.show()
```

````python
import math

import numpy as np
from matplotlib import pyplot

pi = math.pi   #mathモジュールのπを利用

x = np.linspace(0, 2*pi, 100)  #0から2πまでの範囲を100分割したnumpy配列
y = np.sin(x)

pyplot.plot(x, y)
pyplot.show()
````

## グラフの設定と複数描画

````python
import math

import numpy as np
from matplotlib import pyplot

pi = math.pi   #mathモジュールのπを利用

x = np.linspace(0, 2*pi, 100)  #0から2πまでの範囲を100分割したnumpy配列
sin_y = np.sin(x)
cos_y = np.cos(x)


#凡例のためにlabelキーワードで凡例名をつける
pyplot.plot(x, sin_y, label='sin')
pyplot.plot(x, cos_y, label='cos')

#グラフタイトル
pyplot.title('Sin Graph')

#グラフの軸
pyplot.xlabel('X-Axis')
pyplot.ylabel('Y-Axis')

#グラフの凡例
pyplot.legend()

pyplot.show()
````
