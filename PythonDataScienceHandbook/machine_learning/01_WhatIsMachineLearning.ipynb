# 機械学習とは何か
機械学習のメソッドのデータサイエンスでの応用においては、機械学習をデータモデルの構築方法であると考えることがより有用である。

## 機械学習のカテゴリー
基礎的なレベルにおいては機械学習は2つの主要なタイプに分類される。教師あり学習と教師無し学習である。

教師あり学習は、どのようにして測定したデータと、データと関連するラベルとの間の関係性のモデル化するかというものである。いったんこのモデルが決定されると、新しいラベルや未知のデータに対しても応用することが出来る。さらに、これは分類（classification)と回帰(regression)に分けられる。分類ではラベルは不連続なカテゴリーである一方、回帰ではラベルは連続した量である。

教師無し学習はどんなラベルに対しても参照のないデータセットの特徴をモデル化するものである。そしてそれはデータセットに自身について語らせるというように記述される。これらのモデルはクラスタリングや次元削減のようなタスクを含んでいる。クラスタリングのアルゴリズムはデータの明確なグループを識別する一方、次元削減はデータをより簡潔に表現を調べる。

加えて、準教師あり学習と呼ばれるメソッドもある。
