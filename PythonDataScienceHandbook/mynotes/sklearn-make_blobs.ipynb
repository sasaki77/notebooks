# scikit-learnで分類問題用のサンプルを生成する

sklearn.datasets.make_blobsを利用する。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(
                   n_samples=200,  # サンプル数
                   n_features=2,   # サンプルのマトリックスの列数
                   centers=3,      # サンプルの分類数
                   random_state=3, # ランダムデータ生成のためのシード
                   cluster_std=1.5 # サンプルの標準偏差
                 )
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap='viridis');
```
