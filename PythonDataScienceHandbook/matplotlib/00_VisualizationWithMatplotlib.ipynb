## コンテキストごとのプロット方法
### スクリプト
`plt.show()`コマンドによって、イベントループが開始され、ウィンドウが作成される。同じPythonのセッションで複数回`show()`を呼ぶと、予期せぬ動作をする恐れがあるので、避けること。

```python
# ------- file: myplot.py ------
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

### IPython shell
IPythonはMatplotlibモードを指定することで、Matplotlibととても上手く動作するように作られている。このモードは`%matplotlib`というマジックコマンドを使用することで利用できる。このモードでは、`plt`コマンドが図のウィンドウを開き、追加のコマンドは図を更新することが出来る。いくつかの変更は自動的に描画されないが、その場合は`plt.draw()`で強制的に更新できる。`plt.show()`はこの場合不要である。

### Jupyter notebook
IPython shellと同様に`%matplotlib`のマジックコマンドで利用可能。Jupyter notebookでは以下の2つのオプションが利用できる。

- `%matplotlib notebook`: インタラクティブなプロットとしてノートブックに埋め込む
- `%matplotlib inline`: 静的な画像としてノートブックに埋め込む

## 二つのインターフェース

````python
import numpy as np

x = np.linspace(0, 10, 100)
````

### MATLABスタイルインターフェース

````python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
````

### オブジェクト指向インターフェース
基本的にはこっちを使う。

````python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
````
